```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
df= pd.read_csv('diabetes.csv')
df
```

# Split data into input and target features


```python
X=df.drop(columns=['Outcome'], axis=1)
Y= df['Outcome']
```

# Split entire data into 2 subsets
- Subset 1 (Train Set)
- Subset 2 (Validation Set / Hold Out Set)

```python
from sklearn.model_selection import train_test_split
train, val_train, test, val_test= train_test_split(X, Y, test_size=0.5, random_state=42)
```

# Split Train set further into 
- Train Set (80%)
- Test Set (20%)

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Model 1: Implementation of KNeighborsClassifier

```python
knn= KNeighborsClassifier()
knn.fit(X_train, Y_train)
knn.score(X_test, Y_test)
```

# Model 2: Implementation of SVC Classifier

```python
svc= SVC() 
svc.fit(X_train, Y_train)
svc.score(X_test, Y_test)
```

# Predictions of the model on Validation Set

```python
predict_val1= knn.predict(val_train)
predict_val2= svc.predict(val_train)
```

```python
predict_val1
```

```python
predict_val2
```

```python
predict_val =np.column_stack((predict_val1, predict_val2))
predict_val
```

```python
# Prediction of Test set
predict_test1= knn.predict(X_test)
predict_test2= svc.predict(X_test)
predict_test =np.column_stack((predict_test1, predict_test2))
```

```python
predict_test
```

### Stacking: Input features are predictions of Model M1(KNN) and Model M2(SVC) stored inside predict_val (input features) for RandomForestClassiier

```python
from sklearn.ensemble import RandomForestClassifier
rand_clf= RandomForestClassifier()
rand_clf.fit(predict_val, val_test)
```

```python
rand_clf.score(predict_test, Y_test)
```

# Hyperparameter Tuning: GridSearchCV

```python
grid_param ={
    'n_estimators': [90,100,115],
    'criterion': ['gini', 'entropy'],
    'min_samples_leaf': [1,2,3,4,5],
    'min_samples_split': [4,5,6,7,8],
    'max_features': ['auto', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search=GridSearchCV(estimator=rand_clf, param_grid= grid_param, cv=5, n_jobs=-1, verbose=3 )
grid_search.fit(predict_val, val_test)
```

```python
grid_search.best_params_
```

```python
rand_clf_hyperparameter_tuning = RandomForestClassifier(n_estimators= 90,max_features='log2',min_samples_leaf= 1,min_samples_split= 4)
rand_clf_hyperparameter_tuning.fit(predict_val, val_test)
```

```python
rand_clf_hyperparameter_tuning.score(predict_test, Y_test)
```
